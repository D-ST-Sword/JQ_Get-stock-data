```python
import datetime
rows = []
for i in range(8):
    csvrow = get_bars(cod[i], 1200, unit='1d',
             fields=['date','high','low','open','close'],
             include_now=False, end_dt='2020-11-12', fq_ref_date=datetime.date.today(),df=True)
    rows.append(csvrow)
```

```python
#获得的股票历史数据存入文件
import os
import csv
 
```

```python
rowlist=[]
for i in range(8):
    tmp = rows[i].values.tolist()
    rowlist.append(tmp)
```

```python
#这里设置一个文件名抬头list,（也可以在获取股票代码时统一设置好）
namelist=["example0","example1","example2","example3","example4","example5","example6","example7"]
```

```python
#开始存储
for i in range(8):
    
    current_dir = os.path.abspath('.')
    file_name = os.path.join(current_dir, namelist[i] + '.csv')#将存储的文件名补全
    csvfile = open(file_name, 'wt' ,encoding="UTF8") 
    writer=csv.writer(csvfile, delimiter=",")
    header=['date','high','low','open','close']#有多少特征全部分列存储，日期也需要单独一列
    
    da=[]
    hi=[]
    lo=[]
    op=[]
    cl=[]
    for j in range(959):
        print('i',i,'j',j)
        da.append(rowlist[i][j][0])
        hi.append(rowlist[i][j][1])
        lo.append(rowlist[i][j][2])
        op.append(rowlist[i][j][3])
        cl.append(rowlist[i][j][4])
        
    
    writer.writerow(header)
    writer.writerows(zip(da,hi,lo,op,cl))
    csvfile.close()
    

```
